```python
from math import inf

def fw(V, E):
    # Initialization
    n = len(V)
    D = [[inf for _ in range(n)] for _ in range(n)]
    
    # Set the distance from each vertex to itself to 0
    for i in range(n):
        D[i][i] = 0
    
    # Set the initial distances based on the edges
    for (i, j, w) in E:
        D[i-1][j-1] = w
    
    print(f"D{0}")
    print_matrix(D)

    # Floyd-Warshall algorithm
    for k in range(n):
        print(f"D{k+1}")
        for i in range(n):
            for j in range(n):
                if D[i][j] > D[i][k] + D[k][j]:
                    print(f"D[{i}][{j}] changed  from {D[i][j]} to {D[i][k] + D[k][j]}")
                    D[i][j] = D[i][k] + D[k][j]
        
        print_matrix(D)
    
    # return D

def print_matrix(d):
    for row in d:
        print(" ".join(f"{v:3}" if v != inf else "inf" for v in row))
    print()

# Example usage
Vertices = [1, 2, 3, 4]
Edges = [(1, 2, -1), (1, 4, 6), (2, 4, 5), (3, 1, 3), (3, 2, 3), (4, 3, -4)]

fw(Vertices, Edges)
```
